```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
%matplotlib inline
```

```python


def f(x):
    return -np.exp(x) +1

def bisection(a,b,tol):
    xl=a # lower boundary
    xr=b # upper boundary
    while (np.abs(xl-xr)>=tol):
        c=(xl+xr)/2.0
        prod=f(xl)*f(c)
        if prod>tol:
            xl=c
        else:
            if prod<tol:
                xr=c
    return c
answer=bisection(-5,5,1e-8)
print("Bisection Method gives root at x=",answer)
x=np.linspace(-5,5,100)
plt.plot(x,f(x))
plt.grid()
plt.title('f(x) vs x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
