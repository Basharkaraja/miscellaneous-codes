```python
def rho_free(x,x1,beta):
    rho=(1/(np.sqrt(2*math.pi*beta))) *(math.exp((-x-x1)**2)/(2*beta))
    return rho
```

```python
def rho_harmonic(x,x1,beta,k):
    rho_fr=rho_free(x,x1,beta)
    rho_h_o=math.exp(-0.5*k*x**2)*math.exp(-rho_fr*beta)*math.exp(-0.5*k*x1**2)
    return rho_h_o
```
