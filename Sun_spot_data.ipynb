```python
# let's plot all the fit lines frm order 1 to order 10 and see which one is the best fit:
total_fn=[]
for i in range(1,11):
    curve=np.polyfit(year,sun_spot,i)
    polynomial=np.poly1d(curve)

    new_y=polynomial(year)
    total_fn+=[new_y]
    

for i in range(1,10):
    plt.plot(year,sun_spot,'ro',Label='original points')
    plt.plot(year,total_fn[i],Label='order'+str(i))
    plt.show()
    plt.legend()
    plt.title('order'+str(i))
    
```

```python
# bias=expected value of the point -  real value
bias=0
t_variance=[]
bias_of_all_fn=[]
for i in total_fn:
    index=0
    for j in i:
        bias+=j - year[index]
    bias_of_all_fn+=[bias]
    bias=0
    t_variance+=[statistics.variance(i)]
for i in range(len(bias_of_all_fn)):
    print('bias of order',i,'is: ',bias_of_all_fn[i],'and variance: ',str(t_variance[i]))
    
    
```

it is clear here that the polynomial regression model is not accurate, but from all the polynomial orders, we can see that the best fit between them for our current data is the polynomial of order 7, since it has the smallest bias among the other orders

# bootstraping:

we will create a function that takes as an argument: 

1- our data

2- number of times we want to bootstrap the data

3- and the statistics of our data

```python
def bootstrap(data,n_boots,stat):
    data=np.array(data)
    resampled=[]
    
    for i in range(n_boots):
        # we will be randomly chosing an index that will pull out a random sample from our data
        index=np.random.randint(0,len(data),len(data))
        sample=data[index]
        statistics=stat(sample)
        resampled.append(statistics)
        
        
    return np.array(resampled)


```
